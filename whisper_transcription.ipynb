# 🎙️ Speech-to-Text with OpenAI Whisper
Transcribe speech from a video or audio file.

```python
import whisper

# Load model — choose: 'tiny', 'base', 'small', 'medium', 'large'
# Larger = more accurate but slower
model = whisper.load_model('base')
print('Model loaded!')
```

```python
# Set the path to your video or audio file
VIDEO_PATH = 'your_video.mp4'  # <-- change this

# Transcribe
result = model.transcribe(VIDEO_PATH)

# Print the full transcript
print(result['text'])
```

```python
# Save transcript to a text file
with open('transcript.txt', 'w') as f:
    f.write(result['text'])

print('Saved to transcript.txt')
```

```python
# BONUS: Print transcript with timestamps (segment by segment)
for segment in result['segments']:
    start = segment['start']
    end = segment['end']
    text = segment['text']
    print(f'[{start:.1f}s --> {end:.1f}s]  {text}')
```
